```python
def save_esther_profiles_all(creds):
    global df_persons
    global lastload_persons
    
    for id in df_persons.index:
        print id
        person_id = id
        firstname = df_persons.firstname[id]
        properties = df_persons.properties[id]
        segment_scores = df_persons.segment_scores[id]
        conv_state = df_persons.conv_state[id]
        conv_substate = df_persons.conv_substate[id]
        conv_last_update = df_persons.conv_last_update[id]

        if conv_last_update > lastload_persons:
            print id, 'Yes'
            save_esther_profile(creds, person_id, firstname, properties, segment_scores, conv_state, conv_substate, conv_last_update)
```

```python
from numpy.random import rand
import random
import numpy as np
from matplotlib import pyplot as plt
import time
%matplotlib inline

def smooth(binmeans):
    binsums_new = np.zeros(binmeans.shape)
    bin_weights = np.zeros(binmeans.shape)
    for i in range(binmeans.shape[0]):
        for j in range(binmeans.shape[1]):
            for ii in range(max(i-1,0),min(xbins,i+1)):
                for jj in range(max(j-1,0),min(ybins,j+1)):
                    if np.isfinite(binmeans[ii,jj]):
                        binsums_new[i,j] += binmeans[ii,jj]
                        bin_weights[i,j] += 1
    binmeans_new = np.where(bin_weights > 0, binsums_new/bin_weights, np.nan)
                    
    return binmeans_new

NN = 2000


NN1 = 2000
NN2 = 2001
step = 1000


age_days = np.random.rand(NN)*30
#offers_sent = np.array([np.random.randint(0,n) for n in age_days])
offers_sent = np.array([np.random.randint(0,25) for n in age_days])
take_up_prob = 0.2 + 0.6*np.exp(-((age_days - 20)**2 +(offers_sent - 6)**2)/50)
randvect = [random.random() for i in take_up_prob]
taken_up = np.where(take_up_prob > randvect, 1, 0)


for N in range(NN1,NN2,step):

#N = 100



    #plt.show()

    xbins = 30
    ybins = 30

    x = age_days[:N]
    y = offers_sent[:N]
    zprob = take_up_prob[:N]

    
    z = taken_up[:N]
    
    color = np.where(z == 1, 'red', 'blue' )
    
    markers = list(np.where(z == 1, u'x', u'o' ))
    
    fig, ax = plt.subplots()
    #for color in ['red', 'green', 'blue']:
    n = 100
    #x, y = rand(2, n)
    #scale = 200.0 * rand(N)
    scale = 10
    for _x,_y,_c,_m in zip(x,y,color,markers):
        ax.scatter(_x, _y, c = _c, s=scale, label=_c, marker = _m)#,
               #alpha=0.3, edgecolors='none')

    #ax.legend()
    ax.grid(True)
    
    plt.ylabel('Non-health Offers')
    plt.xlabel('Days Since Enrollment')
    plt.grid(True)
    plt.ylim((0,25))
    plt.xlim((0,30))
    #def smooth(xbins, ybins, slope, x, y, z, N):
    #minx = min(x)
    #miny = min(y)
    #maxx = max(x)
    #maxy = max(y)
    minx = 0
    miny = 0
    maxx = 30
    maxy = 25
    x_bin = np.zeros((N,), dtype=np.int)
    y_bin = np.zeros((N,), dtype=np.int)

    bincounts = np.zeros((ybins,xbins), dtype=np.int)
    binsums = np.zeros((ybins,xbins))

    for i in range(N):
        #print i, x[i], y[i], x_bin[i], y_bin[i]
        x_bin[i] = min(int(xbins*(x[i] - minx)/(maxx-minx)), xbins - 1)
        y_bin[i] = min(int(ybins*(y[i] - miny)/(maxy-miny)), ybins - 1)
        #print i, x[i], y[i], x_bin[i], y_bin[i]
        bincounts[y_bin[i], x_bin[i]] += 1
        binsums[y_bin[i], x_bin[i]] += z[i]

    binmeans = binsums/bincounts




    a = binmeans
    #plt.imshow(a, cmap='cool', interpolation='nearest')
    #plt.show()

    #time.sleep(1)

    b = smooth(binmeans)
    #plt.imshow(b, cmap='cool', interpolation='nearest')
    #plt.show()
    #plt.legend()

    for i in range(5):
        b = smooth(b)
        #plt.imshow(b, cmap='cool', interpolation='nearest')
        
        where_are_NaNs = np.isnan(b)
        c = b
        #c[where_are_NaNs] = 0
        
    plt.imshow(b, cmap='cool', interpolation='nearest', origin='lower')
    
    plt.savefig('sweetspot.png', dpi=1000)
    plt.show()
 
    
    
```

```python
def list_best_offers(person_id, num_offers = 10):

    prob_takeup = predict_offer_takeup(person_id, globals.df_persons, globals.model_list_offers, globals.df_offers)

    df_offer_selection = globals.df_offers[['offer_id', 'offer_name']]
    df_offer_selection['prob_takeup'] = prob_takeup[1:]                                    

    df_offer_selection = df_offer_selection.sort_values(by = 'prob_takeup', ascending = False)
    df_offer_selection = df_offer_selection.reset_index()

    send_message(person_id, 'Offer:                   Probablility of Take-up')
    for i in df_offer_selection.index[:num_offers]:
        label = str(df_offer_selection.offer_name[i])
        tu_perc = str(int(100*df_offer_selection.prob_takeup[i]))+'%'
        spaces = ' '*(30-len(label))
        send_message(person_id, label + ': ' + spaces + tu_perc)
```
